## Fitting the model parameters

We'll use Treasury yield curve data to fit the mean reversion parameter, $\alpha$, using Treasury data, assuming that this value
is also relevant for the 1-month CD data.

Define $g$ as:

$$g = \sqrt{\alpha^2 + 2\sigma^2}$$

so that

$$ \frac{dg}{d\alpha} = \frac{\alpha}{g}$$

```python
def g_dg(alpha, sigma):
    g = math.sqrt(alpha**2 + 2 * sigma ** 2)
    dg = alpha / g
    return g, dg
```

The we get the following components of the affine yield (spot) curve implied by the model:

$$A(T; \alpha, b, \sigma) = \frac{2\left(e^{\ gT} - 1\right)}{(g+\alpha)(e^{\ gT} - 1) + 2g}$$

```python
def A(alpha, b, sigma, T):
    g, _ = g_dg(alpha, sigma)
    num = 2 * (np.exp(g * T) - 1)
    den = ((g + alpha) * (np.exp(g * T) - 1) + 2 * g)
    a = num / den
    return a
```

and, 

$$ C(T; \alpha, b, \sigma) = \frac{2\alpha b}{\sigma^2}\log\left\[\frac{2ge^{\ (\alpha + g)T / 2}}{(\alpha + g)(e^{\ gT} -1) + 2g}\right\] $$

```python
def C(alpha, b, sigma, T):
    g, _ = g_dg(alpha, sigma)
    num = 2 * g * np.exp((alpha + g) * T / 2.)
    den = (g + alpha) * (np.exp(g * T) - 1) + 2 * g
    c = (2 * alpha * b / (sigma**2)) * np.log(num / den)
    return c
```

The affine yield formula is then:

$$Y(T, r_0; \alpha, \beta, \sigma) = [A(T; \alpha, b, \sigma)r_0 - C(T; \alpha, b, \sigma)]/T$$

```python
def Y(alpha, b, sigma, T, r0):
    y = (A(alpha, b, sigma, T) * r0 - C(alpha, b, sigma, T) ) / T
    return y
```

Define the sum-of-squared errors loss function as

$$ L(T, y, r_0; \alpha, b, \sigma) = .5\times \sum\left\[y_i - Y(T_i, r_0; \alpha, b, \sigma)\right\]\ ^{\ 2} $$

where $y_i$ are the observed spot rates at each tenor $T_i$.

```python
def cir_loss(alpha, b, sigma, T, r0, y_obs):
    
    y_pred = Y(alpha, b, sigma, T, r0)
    err = y_obs - y_pred
    loss = .5 * np.dot(err, err)
    
    return loss
```

```python
alpha0 = .8
alpha_sol = opt.fmin_bfgs(cir_loss, alpha0, fprime = None, args = (tr_lt, cd_vol, tenors[1:], tr0, spots[1:]))
print "Solution (alpha) = ", alpha_sol
```

#### Plot the observed spot rates vs. the model fit.

The model fit is bound to be relatively poor, because our model is very simple and constrained in all but one parameter.

This is okay, because we prefer to:

1. Rely on historical data, since this is an actuarial application, not an asset pricing one.
2. Keep the model fitting procedure simple-to-implement and likely-to-converge.

```python
plt.plot(tenors, Y(alpha_sol, tr_lt, cd_vol, tenors, tr0), label = 'Model')
plt.plot(tenors, spots, 'ob', label = 'Actual')
plt.legend(loc = 'best')
```

#### Graph the loss function and check that we've achieved a minimum

```python
alpha_range = alpha_sol + np.linspace(-.5, .5, 500)
ll = np.array([cir_loss(a, tr_lt, cd_vol, tenors[1:], tr0, spots[1:]) for a in alpha_range])

plt.plot(alpha_range, ll)
plt.axvline(alpha_sol)
```

## Simulating future rates

The Cox-Ingersoll-Ross model of the short rate is a square root diffusion process of the form
$$dr(t) = \alpha(b - r(t))\ dt + \sigma\sqrt{r(t)}\ dW(t).$$

We wish to simulate the process at discrete time points, $t_0, t_1, \ldots, t_n\ $. To do so, it's helpful
to define the following variables:

$$(1)\qquad  \Delta_i = t_{i+1} - t_i \qquad i = 0, \ldots, n-1$$

$$(2)\qquad d = 4b\alpha/\sigma^2 $$

$$(3)\qquad c = \sigma^2\left(1-e^{\ \alpha\ \Delta_i}\right)/4\alpha $$

$$(4)\qquad \lambda = r(t_i)\left(e^{\ \alpha \Delta_i}\right)/c$$



```python
def sim_cir(time_vec, r0, alpha, b, sigma):
    n = len(time_vec)
    d = 4 * b * alpha / (sigma**2)
    dtime = time_vec[1:] - time_vec[:-1]
    c = (sigma**2) * (1 - np.exp(-alpha * dtime)) / (4 * alpha)
    rates = np.zeros(n)
    rates[0] = r0
    
    if d > 1:
        Z = np.random.normal(0, 1, n)
        X = np.random.chisquare(int(d-1), n)
        for i in xrange(n-1):
            lam = rates[i] * (math.exp(-alpha * dtime[i]) / c[i])
            rates[i+1] = c[i] * ((Z[i] + sqrt(lam))**2  + X[i])
    else:
        for i in xrange(n-1):
            lam = rates[i] * (math.exp(-alpha * dtime[i]))/ c[i]
            N = np.random.poisson(int(lam / 2.), 1)
            X = np.random.chisquare(int(d + 2 * N), 1)
            rates[i+1] = c[i] * X
    return(rates)                 
```

### Plot some sample paths, and the average

```python
months = arange(1, 133)
times = months / 12.
nsims = 500
rr_mean = np.zeros_like(times)
for i in xrange(nsims):
    rr = sim_cir(times, cd0, alpha_sol, cd_lt, cd_vol)
    rr_mean = rr_mean + rr
    plt.plot(times, rr, '-k', alpha = .05)
plt.plot(times, rr_mean / nsims, '-r', lw = 3, label = 'Mean Path')
plt.xlabel('years')
plt.legend(loc = 'best')
```

## Simulate the 1-month rates for the next 11 years.

```python
# Simulation parameters
nsims = 10000
months = arange(0, 134)
times = months / 12.
# Results container
sim_matrix = np.zeros((nsims, len(times)))
```

```python
for i in xrange(nsims):
    sim_matrix[i, :] = sim_cir(times, cd0, alpha_sol, cd_lt, cd_vol)
```

#### Plot mean and ranges of the simulations, compared with historical data

```python
past_years = 10
past_rates = cd1m['cd1m'][-past_years * 12:]
plt.figure(figsize = (10, 6))
plt.plot(arange(-len(past_rates)+1, 1) / 12., past_rates)
plt.plot(times, np.mean(sim_matrix, axis = 0), '-k')
plt.fill_between(times, np.min(sim_matrix, axis = 0), 
                        np.max(sim_matrix, axis = 0), 
                 color = 'red', alpha = .5)

plt.fill_between(times, np.percentile(sim_matrix, 2.5, axis = 0), 
                        np.percentile(sim_matrix, 97.5, axis = 0),
                 alpha = 1)
plt.xlim(-past_years, 11)

plt.title(('''Past cash rates and simulated future rates
            (Average path with min-to-max, and 95% confidence ranges)'''))
plt.xlabel('Time (years)')
plt.ylabel('Annualized rate (%)')
plt.savefig(os.path.join(graphdir, 'Cash simulation chart.pdf')) 
```

### Make the simulation results a dataframe, and export it.

```python
sim_results = DataFrame(sim_matrix[:, 1:] / 1200., 
                        columns = ['month' + str(i) for i in months[1:]],
                        index = arange(nsims) + 1)
sim_results.index.name = 'simulation'
```
