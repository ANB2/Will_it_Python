```python
import cylowess as cylimport statsmodels.nonparametric.lowess as smlwfrom pandas import *
```

```python
################################################# Comparison of fits:# -------------------## I use the Silverman motorcycle data to# visually compare the results of statsmodels# lowess, new lowess, and R's lowess.# # The same defaults are used in all three:# frac = 0.1 (so k = int(0.1 * n)# 3 robustness iters (after the inital fit)# no `delta` adjustment in R or new lowess################################################
```

```python
# The Silverman motorcycle data, exported from R's MASS package.# time is the number of milliseconds after impact.# accel is the acceleration of the vehicle.mcycle = read_csv('mcycle.csv')time = mcycle['times']accel = mcycle['accel']
```

```python
# Note that the new lowess takes one more `it` than the original,# since the first `it` is the initial fit. This is consistent with# R.sm_lowess = smlw.lowess(accel, time, frac = 0.1, it = 3)new_lowess = cyl.lowess(accel, time, frac = 0.1, it = 4)# Results from R: lowess(mcycle, delta = 0, f = 0.1, iter =4)r_lowess = read_csv("r_lowess_d0_it4_f0-01.csv")
```

```python
plt.figure(figsize = (10, 7))plt.plot(time, accel, '.', color = "steelblue", alpha = 0.25, label = 'Data')plt.xlabel('Time after impact (ms)')plt.ylabel('Acceleration (g)')plt.plot(new_lowess[:,0], new_lowess[:,1], '-', color = 'orange', label = 'New lowess')plt.plot(sm_lowess[:,0], sm_lowess[:, 1], '--r', label = 'Statsmodel lowess')plt.plot(r_lowess['x'], r_lowess['y'], '+g', label = 'R lowess')plt.legend(loc = 'upper left')
```

```python
############################################# Timings:# --------# With 10,000 points, at the default# frac, statsmodels lowess is quite slow# (around 25 seconds on my machine).# The new lowess is substantially faster,# and comparable with R (9-11 seconds vs. # 5.1 in R). # Using the delta parameter timings fall# to the 100s of milliseconds.## Note the timings are done at the default# frac of 2/3, which is not a good frac for# this data. This is just done for timing# purposes. To compare output from stats-# models lowess and new lowess with delta,# I used a more appropriate, smaller, frac.# Both methods are faster with this frac.# (About 8 seconds for statsmodels, 3.5#  for new lowess w/o delta)############################################
```
